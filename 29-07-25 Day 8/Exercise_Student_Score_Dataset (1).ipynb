```python
import pandas as pd

# Read the students.csv file
df = pd.read_csv("/content/sample_data/students.csv")

# Print the first 3 rows
print("First 3 rows of the dataset:")
print(df.head(3))
```

```python
# Add new column 'Average' for each student's mean score
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)

# Print the student with the highest average
top_student = df.loc[df["Average"].idxmax()]
print("Student with the highest average:")
print(top_student)
```

```python
# Students who scored more than 80 in science
print("students who scored more than 80 in science: ")
print(df[df["Science"]>80])
```

```python
#Students with average below 70
print("\nStudents with average below 70:")
print(df[df["Average"]<70])
```

```python
#sort students by english score in descending order
sorted_df = df.sort_values(by="English", ascending=False)
#print top 3 students by english score
print("Top 3 students by English score:")
print(sorted_df.head(3))
```

```python
import matplotlib.pyplot as plt

#Bar chart: Students name vs average score
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Average"],color="skyblue")
plt.title("Average Score per Student")
plt.xlabel("Students Name")
plt.ylabel("Average Score")
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()

#Line chart: scores in all subjects
df.set_index("Name")[["Math", "Science", "English"]].plot(marker='o', figsize=(8, 5))
plt.title("Scores in All Subjects")
plt.xlabel("Students Name")
plt.ylabel("Scores")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Find subject with the highest total score across all students.
subject_totals = df[["Math", "Science", "English"]].sum()
top_subject = subject_totals.idxmax()
top_score = subject_totals.max()

print(f"Subject with the highest total score: {top_subject} {top_score}")

# Save the updated DataFrame to a new CSV file
df.to_csv("students_summary.csv", index=False)
print("Updated DataFrame saved as students_summary.csv")

```
